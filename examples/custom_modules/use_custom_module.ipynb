# Custom Modules in TerraTorch

This notebook demonstrates how to extend TerraTorch using the custom modules pattern. This workflow allows you to register user-defined backbones (like HelloGeoModule) into the internal registry, enabling you to swap architectures via simple string configurations without modifying the core library.

```python
import sys
from lightning.pytorch import Trainer

from terratorch.tasks.regression_tasks import PixelwiseRegressionTask
from terratorch.models.decoders.identity_decoder import IdentityDecoder
from terratorch.datamodules.dummy_datamodule import DummyDataModule

datamodule = DummyDataModule(
    batch_size=4,
    num_workers=2,
    train_samples=20,
    val_samples=10,
    test_samples=10,
    height=256,
    width=256,
)

task = PixelwiseRegressionTask(
    model_factory='EncoderDecoderFactory',
    model_args={
        "backbone": "HelloGeoModule",  
        "backbone_input_channels": 3,
        "backbone_output_channels": 64,
        "backbone_kernel_size": 3,
        "decoder": 'IdentityDecoder',
    },
    loss="mse",
    lr=0.001,
)

trainer = Trainer(
    accelerator="auto",
    devices=1,
    max_epochs=2,
    log_every_n_steps=5,
)

trainer.fit(task, datamodule=datamodule)
```
